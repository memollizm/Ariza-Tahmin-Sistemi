```python
from sklearn.manifold import TSNE

features = data[['Hava SÄ±caklÄ±ÄŸÄ± [K]', 'Ä°ÅŸlem SÄ±caklÄ±ÄŸÄ± [K]', 'DÃ¶nme HÄ±zÄ± [rpm]', 'Tork [Nm]']]
tsne = TSNE(n_components=2)
tsne_result = tsne.fit_transform(features)

plt.figure(figsize=(10, 6))
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=data['Makine ArÄ±zasÄ±'], cmap='viridis')
plt.colorbar(label='Makine ArÄ±zasÄ±')
plt.title('t-SNE Verilerin GÃ¶rselleÅŸtirilmesi')
plt.show()
```
