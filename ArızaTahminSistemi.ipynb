```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('ai4i2020.csv')
```

```python
turkish_columns = {
    'UDI': 'Benzersiz Kimlik Numarası',
    'Product ID': 'Ürün Kimliği',
    'Type': 'Tip',
    'Air temperature [K]': 'Hava Sıcaklığı [K]',
    'Process temperature [K]': 'İşlem Sıcaklığı [K]',
    'Rotational speed [rpm]': 'Dönme Hızı [rpm]',
    'Torque [Nm]': 'Tork [Nm]',
    'Tool wear [min]': 'Takım Aşınması [dk]',
    'Machine failure': 'Makine Arızası',
    'TWF': 'Takım Aşınması Arızası',
    'HDF': 'Yüksek Hızlı Mil Arızası',
    'PWF': 'Güç Arızası',
    'OSF': 'Soğutucu Akışkan Arızası',
    'RNF': 'Sensör Arızası'
}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numeric_df = data.select_dtypes(include=['float64', 'int64'])
```

```python
corr_matrix = numeric_df.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korelasyon Matrisi')
plt.show()
```

```python
plt.scatter(data['Dönme Hızı [rpm]'],data['Tork [Nm]'])
plt.xlabel('Dönme Hızı [rpm]')
plt.ylabel('Tork [Nm]')
plt.title('Dönme Hızı ve Tork Arasındaki İlişki')
plt.show()
```

```python
sns.boxplot(x='Tip', y='Hava Sıcaklığı [K]', data=data)
plt.title('Ürün Tipine Göre Hava Sıcaklığı Dağılımı')
plt.show()
```

```python
sns.jointplot(x='Tip', y='Tork [Nm]', data=data, kind='scatter', color='red')
plt.show()
```

```python
sns.violinplot(x='Tip', y='Takım Aşınması [dk]', data=data, color='purple')
plt.title('Ürün Tipine Göre Takım Aşınması')
plt.show()
```

```python
sns.histplot(data=data, x='Tip', y='Güç Arızası', color='Red')
plt.title('Ürün Tipine Göre Güç Arızası')
plt.show()
```

```python
sns.jointplot(data=data, x='Tip', y='Sensör Arızası')
plt.title('Ürün Tipine Göre Sensör Arızası')
plt.show()
```

```python
sns.displot(data, x="Tork [Nm]", y="İşlem Sıcaklığı [K]", kind="kde")
plt.title("İşlem Sıcaklığının Tork'a Etkisi")
plt.show()
```

## Veriler Arasındaki İlişkilerin Daha Detaylı İncelenmesi

```python
sns.barplot(data=data, x="Güç Arızası", y="Tork [Nm]", color="green")
```

```python
sns.pairplot(data, vars=["Hava Sıcaklığı [K]", "İşlem Sıcaklığı [K]", "Dönme Hızı [rpm]", "Tork [Nm]", "Takım Aşınması [dk]"], hue="Makine Arızası")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data["Dönme Hızı [rpm]"], data["İşlem Sıcaklığı [K]"], data["Tork [Nm]"], depthshade=True)
```

```python
pivot_table = data.pivot_table(values='Tork [Nm]', index='Tip', columns='Makine Arızası', aggfunc='mean')
sns.heatmap(pivot_table, annot=True, cmap='viridis')
```

```python
from sklearn.manifold import TSNE

features = data[['Hava Sıcaklığı [K]', 'İşlem Sıcaklığı [K]', 'Dönme Hızı [rpm]', 'Tork [Nm]']]
tsne = TSNE(n_components=2)
tsne_result = tsne.fit_transform(features)

plt.figure(figsize=(10, 6))
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=data['Makine Arızası'], cmap='viridis')
plt.colorbar(label='Makine Arızası')
plt.title('t-SNE Verilerin Görselleştirilmesi')
plt.show()
```
